```python
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv("Train_Data.csv")
df_train
```

```python
df_test = pd.read_csv("Test_Data.csv")
df_test
```

```python
df_train.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train = df_train.drop(['attack'], axis='columns')
y_train = df_train['attack']
```

```python
y_train
```

```python
x_test = df_test
x_test
```

```python
# Encoding categorical features
categorical_features = ['protocoltype', 'service', 'flag']
numerical_features = x_train.columns.difference(categorical_features)
```

```python
numerical_features
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),  # Apply StandardScaler to numerical features
        ('cat', OneHotEncoder(), categorical_features)  # Apply OneHotEncoder to categorical features
    ])
```

```python
# Create pipeline
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier())
])
```

```python
# Split data for validation
x_train_split, x_val_split, y_train_split, y_val_split = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

```python
pipeline.fit(x_train_split, y_train_split)
```

```python
# Validate model
y_val_pred = pipeline.predict(x_val_split)
y_val_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
print(classification_report(y_val_split, y_val_pred))
print("Validation Accuracy: ", accuracy_score(y_val_split, y_val_pred))
```

```python
# Train on full training data
pipeline.fit(x_train, y_train)
```

```python
# Predicting on test data
predictions = pipeline.predict(x_test)
result=predictions
result
```

```python
from sklearn.metrics import f1_score, recall_score, precision_score
```

```python
# converting the result array into a dataframe
results = pd.DataFrame(result, columns=['Result'])
```
